# DataFrame
DataFrame 은 pandas 의 2차원 데이터타입을 말합니다.

```python
data = {'Name' : 'Seoyoon',
        'Age' : 25,
        'Height' : 180.2,
        'Weight' : 60.8,
        'Job' : 'Student'}
data
```

```python
columns = ['Name', 'Age', 'Height', 'Weight', 'Job']
```

```python
# index는 임의로 0으로 두었다.
df = [pd.DataFrame(data = data, index = [0], columns = columns)]
df
```

```python
multi_data = {'Name' : ['Seoyoon', 'Hyunmin'],
        'Age' : [25, 26],
        'Height' : [180.2, 168.7],
        'Weight' : [60.8, 54.3],
        'Job' : ['Student', 'Job seeker']}
multi_data
```

```python
# index는 임의로 0으로 두었다.
multi_df = [pd.DataFrame(data = multi_data, index = range(0,2), columns = columns)]
multi_df
```

# 직접 어떤 식을 데이터프레임에 적용하고 싶을 때

```python
def norm(x):
    return x.max() - x.min()
df.apply(norm)
```

```python
# 위 블럭과 똑같음
df.apply(lambda x: x.max() - x.min())
```
